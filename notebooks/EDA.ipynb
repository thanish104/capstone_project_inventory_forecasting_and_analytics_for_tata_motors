## Which 20% of the spare part contributes 80 % of the sales

```python
# Grouping data by 'Part Desc' and summing 'Spares Amount'
spare_part_sales = (
    df.groupby("Part No")["Spares Amount"]
    .sum()
    .reset_index()
    .rename(columns={"Spares Amount": "Total_Sales"})
)

# Sorting by Total Sales in descending order
spare_part_sales = spare_part_sales.sort_values(by="Total_Sales", ascending=False).reset_index(drop=True)

# Calculating cumulative sales and cumulative percentage
spare_part_sales["Cumulative_Sales"] = spare_part_sales["Total_Sales"].cumsum()
spare_part_sales["Cumulative_Percentage"] = (
    spare_part_sales["Cumulative_Sales"] / spare_part_sales["Total_Sales"].sum()
) * 100

# Finding the 20% of spare parts contributing to 80% of sales
top_20_percent_parts = spare_part_sales[spare_part_sales["Cumulative_Percentage"] <= 80]

# Displaying results
print("Top 20% spare parts contributing to 80% of sales:")
print(top_20_percent_parts)
```

```python
# Plotting the Pareto chart
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar plot for total sales
sns.barplot(
    y="Part No", 
    x="Total_Sales", 
    data=spare_part_sales, 
    palette="Blues_r", 
    ax=ax1
)
ax1.set_ylabel("Total Sales", fontsize=12)
ax1.set_xlabel("Part Description", fontsize=12)
ax1.tick_params(axis="x", rotation=45)

# Line plot for cumulative percentage
ax2 = ax1.twinx()
sns.lineplot(
    y="Part No", 
    x="Cumulative_Percentage", 
    data=spare_part_sales, 
    color="red", 
    marker="o", 
    ax=ax2
)
ax2.set_ylabel("Cumulative Percentage", fontsize=12)
ax2.axhline(80, color="green", linestyle="--", linewidth=1.5, label="80% Threshold")
ax2.legend(loc="upper left")

# Title and layout
plt.title("Pareto Analysis of Spare Part Sales", fontsize=14)
plt.tight_layout()
plt.show()
```
